# Data preprocessing

```python
# Encoding 2 Categorical'lable values' variables {Item, Area} to be numeric using one hot encoding
# create a binary column for each category and returns a matrix with the results

from sklearn.preprocessing import OneHotEncoder
```

```python
yield_df_encod=pd.get_dummies(yield_df,columns=['Area','Item'],prefix =['Country','Item'])
features = yield_df_encod.loc[:,yield_df_encod.columns!='hg/ha_yield']
lable=yield_df['hg/ha_yield']
features.head()

```

```python
features = features.drop(['Year'],axis=1)
features.head()
```

```python
features.info()
```

```python
# feature scaling to get all get all features of the same of level of magnitudes

from sklearn.preprocessing  import MinMaxScaler
scaler = MinMaxScaler()
features=scaler.fit_transform(features)
```

# Training data

```python
# 70/30 for train/test
from sklearn.model_selection import train_test_split
train_data,test_data,train_targets,test_targets = train_test_split(features,lable,test_size=0.3,random_state=42)
# write final df to csv file
#yield_df.to_csv('yield.df')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import svm
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
def compare_models(model):
    model_name = model.__class__.__name__
    fit= model.fit(train_data,train_targets)
    y_pred=fit.predict(test_data)
    r2 = r2_score(test_targets,y_pred) # shows how well terms or datapoints fit a curve 
    return([model_name,r2,])

models = [GradientBoostingRegressor(n_estimators=200, max_depth=3,random_state=0),
          RandomForestRegressor(n_estimators=200,max_depth=3,random_state=0),svm.SVR(),
          DecisionTreeRegressor()]
```
